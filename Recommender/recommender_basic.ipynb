```python
sim15_ratings = pd.merge(sim15, ratings, on="userId").drop(columns=['timestamp'])
sim15_ratings
```

##### sim15_rating is a dataframe of the top 15 users with closest similarity to user 1 and what they've rated other movies, below the weighted rating is being calculated by multiplying the similarty to the users rating, finally the possible rating for user 1 is calculated by first finding the sum of all ratings for a movie and dividing it by the sum of the similarty to all users who rated that movie

```python
sim15_ratings['weighted_rating'] = sim15_ratings['sim_to_user']*sim15_ratings['rating']
```

```python
sim15_ratings = sim15_ratings.groupby('movieId').sum()
sim15_ratings
```

```python
sim15_ratings['possible_rating'] = sim15_ratings['weighted_rating']/sim15_ratings['sim_to_user']
sim15_ratings
```

```python
sim15_ratings = sim15_ratings.merge(movies, on = 'movieId')[['movieId','title','possible_rating']]
sim15_ratings[sim15_ratings['possible_rating'] > 4.49]
```

##### Above is a dataframe of all the movies witha 4.5 or higher possible rating for user 1

```python
user_not_watched = user_not_watched.merge(sim15_ratings, on = ['movieId','title'])
```

```python
user_not_watched['avg_poss_rating'] = user_not_watched[['rating', 'possible_rating']].mean(axis=1)
```
